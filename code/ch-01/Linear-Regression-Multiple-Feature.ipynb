# Linear Regression with multiple feature

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
m = 10 # Training exmaples
x0 = np.ones(m).reshape((m, 1))

# Assuming three features
x = np.random.rand(m,2)
x = np.concatenate((x0, x), axis=1)

# true model: y = 4 + 6 x1 − x2 + e
w = np.array([4 , 6, -1]) 
y = np.dot(x, w) + 0.5 * np.random.randn(10)

cost=[]
n = x.shape[1] # no of features

# initial values for parameters
theta = np.ones(n)
#learning rate
alpha = 0.1 

for iter in range(0, 1000):
    hypothesis = np.dot(x, theta)
    loss = hypothesis - y
    J = np.sum(loss ** 2) / (2 * m)
    gradient = np.dot(np.transpose(x), loss) / m
    theta = theta - alpha*gradient
    cost.append(J)

plt.plot(cost, 'b-')
plt.xlabel('iteration', fontsize = "18")
plt.ylabel('loss', fontsize = "18")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
print('True theta:', w)
print('Predicted theta:', theta)
```
